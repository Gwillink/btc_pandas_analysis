Lets create timeseries from the DataFrame, set the date as index and delete unixtime column.

```python
ts=df.set_index("datetime")
del ts["unixtime"]
```

```python
ts.head(10)
```

between_time: intraday time range

```python
morning_coffee = ts.price.between_time('7:00', '8:00')
```

### Calculations

```python
ts["pv"]=ts["price"]*ts["amount"]
```

```python
ts.pv.plot()
```

```python
ts.pv.sum()
```

```python
ts.pv.max()
```

```python
ts.pv.idxmax()
```

Computing volume-weighted average price (vwap):

```python
vwap = (ts.price * ts.volume).sum() / ts.volume.sum()
vwap
```

###resample: 
regularization and frequency conversion

```python
per_hour = ts.resample("H", how="sum")
per_hour
```

```python
per_hour.volume.plot(kind="bar")
```
